```python
# Importing thr SQL and required pakages
import mysql.connector
from tabulate import tabulate
from datetime import datetime

# building the connection between MYSQL and PYTHON JUPYTER notebook
connection = mysql.connector.connect(
    host = 'localhost',
    user = 'root',
    password = 'root',
    database = 'hospital_database'
)
# Check the connection is connected before performing the 
if connection.is_connected():
    cursor = connection.cursor()
    print("The connection is established")

def rec_appointment():
    print("The Doctor Appointments: ")
    
    try:
        # to display all appointments
        show = "select * from appoint where date(doa) = curdate();"
        cursor.execute(show)
        r = cursor.fetchall()
        print(tabulate(r,headers=["opdno","doa","pname","dname","symptom","treatment"],tablefmt = "pretty"))
        
        # search by the doctor name
        dname = input("Enter Doctor name: ").lower()       
        query = "select opdno, doa,pname, dname, symptom, treatment from appoint where lower(dname) like %s and date(doa) = curdate();"
        cursor.execute(query, ('%' + dname + '%',))
        results = cursor.fetchall()
        
        if results:
            print(tabulate(results, headers=["opdno", "doa", "pname", "dname", "symptom", "treatment"], tablefmt="pretty"))
        else:
            print("No records for the doctor")
            return
            
        # Search by the patient name        
        pname = input("Enter patient name: ").lower()
        query = "SELECT opdno, doa, pname, dname, symptom, treatment FROM appoint WHERE lower(pname) like %s and date(doa) = curdate();"
        cursor.execute(query, ('%' + pname + '%',))
        results = cursor.fetchall()

        if results:
            print(tabulate(results, headers=["opdno", "doa", "pname", "dname", "symptom", "treatment"], tablefmt="pretty"))
        else:
            print("No records for the patient")
         
        # Updating the treatment column
        treatment = input("Treatment: ").lower()

        #sql query to update
        query = "update appoint set treatment = %s where lower(pname) like %s and lower(dname) like %s and date(doa) = curdate();"
            
        cursor.execute(query,(treatment, '%' + pname + '%', '%'+ dname + '%'))
            
        connection.commit()
        
        print("successfully updated treatment.")
       
    except ValueError as val_err:
        print(f"Invalid input:{val_err}")
        
    except mysql.connector.Error as err:
        print(f"Mysql Error is: {err}")
        
    except Exception as e:
        print(f"unknown error is: {e}")
        

def rec_appointmentedit():
    print("Edit or Update the appoinment details: ")
    
    try:
        
        # to display all appointments
        
        show = "select * from appoint where date(doa) = curdate();"
        cursor.execute(show)
        r = cursor.fetchall()
        print(tabulate(r, headers = ["opdno","doa","pname","dname","symptom","treatment"], tablefmt = "pretty"))
        
        
#         show = "select * from appoint"
#         cursor.execute(show)
#         r = cursor.fetchall()

#         print(tabulate(r, headers=["opdno", "doa", "pname", "dname", "symptom", "treatment"], tablefmt="pretty"))
        
        dname = input("Enter Doctor name: ").lower()
        
        query = "select opdno, doa, pname, dname, symptom, treatment from appoint where lower(dname) like %s and date(doa) = curdate();"
        cursor.execute(query,('%' + dname + '%',))
        results = cursor.fetchall()
        
#         query = "select dname from appoint where lower(dname) like %s"
#         cursor.execute(query, ('%' + dname + '%',))
#         r = cursor.fetchone()
        
#         # clear any unread results
#         cursor.fetchall()
        
        if results:
            print(tabulate(results, headers =["opdno","doa","pname","dname","symptom","treatment"], tablefmt = "pretty"))
        else:
            print("No results found: ")
        
         # Search by the patient name        
        pname = input("Enter patient name: ").lower()
        query = "SELECT opdno, doa, pname, dname, symptom, treatment FROM appoint WHERE lower(pname) like %s and date(doa) = curdate();"
        cursor.execute(query, ('%' + pname + '%',))
        results = cursor.fetchall()

        if results:
            print(tabulate(results, headers=["opdno", "doa", "pname", "dname", "symptom", "treatment"], tablefmt="pretty"))
        else:
            print("No records for the patient")
            
        
        # search by treatment
        treatment = input("enter treatment: ").lower()
            
        #update query for the treatment column only
        query = "update appoint set treatment = %s where lower(dname) like %s"
        values = (treatment, '%' + dname + '%')
        cursor.execute(query, values)
        connection.commit()
            
        print("successfully updated treatment.")
                   

    except ValueError as val_err:
        print(f"error is:{val_err}")
        
    except mysql.connector.Error as err:
        print(f"Mysql Error is: {err}")                  
   
    except Exception as e:
        print(f"the unknown error is: {e}")
```
