```python
# Importing thr SQL and required pakages
import mysql.connector
from tabulate import tabulate

# building the connection between MYSQL and PYTHON JUPYTER notebook
connection = mysql.connector.connect(
    host = 'localhost',
    user = 'root',
    password = 'root',
    database = 'hospital_database'
)
# Check the connection is connected before performing the 
if connection.is_connected():
    cursor = connection.cursor()
    print("The connection is established")
    
    
def rec_adddoc():
    print("Add the details of doctor: ")
    
    try:
        did = int(input("enter the doctor ID: "))
        
        check_query = "select * from doctor where did = %s"
        cursor.execute(check_query,(did,))
        results = cursor.fetchone()
  
        if results:
            print(f"doctor with ID {did} already exists")
        
        else:
            dname = input("enter doctor name: ")
            department = input("enter department name: ")
        
            
            #insert the new doctor details
            insert_query = "insert into doctor(did,dname,department) values(%s, %s, %s)"
            cursor.execute(insert_query,(did,dname, department))
        
            connection.commit()
            print("successfully added")
        
            #display the updated list of doctors
            query = "select * from doctor"
            cursor.execute(query)
            result = cursor.fetchall()
        
            table = tabulate(result, headers = ["did","dname","department"],tablefmt = "grid")
            print(table)
     
    
    except ValueError as val_err:
        print(f"Invalid input: {val_err}")
        
    except mysql.connector.Error as err:
        print(f"Mysql Error is: {err}")   
        
    except Exception as e:
        print(f"unknown error is: {e}")     
        
def rec_editdoc():
    print("Updating the doctor details: ")
    
    try:
        # Get the input for the doctor ID
        
        did = int(input("enter the updated did to enter: "))
        
        # check if the doctor ID exists before editing
        
        check = "select * from doctor where did = %s" 
        cursor. execute(check,(did,))
        res = cursor.fetchone()
        
        if res:
            # get new details for the doctor
            dname = input("enter the new name of doctor: ")
            
            department = input("enter the new department name: ")
            
            #update the doctor details
            query = "update doctor set dname = %s, department = %s where did = %s"
            values = (dname, department, did)
            
            # execute the SQL query
            cursor.execute(query,values)
            
            #commit the changes  to the database
            
            connection.commit()
            
            print("Updated successfully")
#             rec_selectdoc()
        else:
            print("Invalid entry. No doctor found with the given ID.")
    except ValueError as val_err:
        print(f"Invalid input: {val_err}")
    
    except mysql.connector.Error as err:
        print(f"Mysql Error is: {err}")
    except Exception as e:
        print(f"the unknown error is:{e}")
        
def rec_selectdoc():
    print("The Doctor Details: ")
    
    try:
        
        query = "select did, dname, department from doctor"
        cursor.execute(query)
        results = cursor.fetchall()
        
        if not results:
            print("no doctors available.")
            return
        headers = ["did","dname","department"]
        table = tabulate(results, headers = headers, tablefmt = "grid")
        print(table)
        
        # ask for the user input
        did = int(input("enter the doctor ID: "))
        
        query = "select * from doctor where did = %s"
        cursor.execute(query,(did,))
        res = cursor.fetchall()
        
        if res:
            headers= ["did","dname","department"]
            values = results
            print(tabulate(res, headers= headers, tablefmt = "pretty"))
   
        else:
            print("no records for the selected doctor ID.")
       
    except ValueError as val_err:
        print(f"Invalid input: {val_err}")
    except mysql.connector.Error as err:
        print(f"Mysql Error is: {err}")
    except Exception as e:
        print(f"unknown error is: {e}")

        
def rec_deletedoc():
    print("Delete doctor details: ")
    try:
        did = int(input("enter the doctor id to delete: "))
        
        # please check if the ID exists before deleting
        check = "select * from doctor where did = %s"
        cursor.execute(check,(did,))
        res = cursor.fetchone()
              
        if res:
            #delete the doctor with the specified ID
            query = "delete from doctor where did = %s"
            cursor.execute(query,(did,))
            connection.commit()
            
            print("deleted successfully")
            
            #display the datails of the deleted doctor
                
            table = tabulate([res], headers = ["dname","department","did"],tablefmt = "pretty")
            print(table)
            
        else:
            print("invalid entry")
    except ValueError as val_err:
        print(f"Invalid input: {val_err}")
    except mysql.connector.Error as err:
        print(f"Mysql Error is: {err}")        
    except Exception as e:
        print(f"unknown error is: {e}")
```
